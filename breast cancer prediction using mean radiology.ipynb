```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
```

```python
data=pd.read_csv(r'C:\Users\sairam\Downloads\breast-cancer.csv')
data.head()
```

The Id column is not needed, hence we can drop it.

```python
data=data.drop(['id'],axis=1)
data.head()
```

```python
data.info()
```

The data does not contain any null values. hence we need to worry about them. The next we can check for duplicate values. 

```python
data.duplicated()
data=data.drop_duplicates()
data.info()
```

The data doesn't contain any duplicate values. Hence we need to worry about them. The next step is to convert the character values present in the diagnosis column to integers so that they can be used as target varaibles for classification. 

```python
data['diagnosis'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
data['diagnosis']=encode.fit_transform(data['diagnosis'])
data['diagnosis'].unique()
data.head()
```

The next step us feature scaling. I thought of bringing all the values to the same scale, hence I standardised the data. 

```python
data1=data.copy()
data1=data1.drop(['diagnosis'],axis=1)
#lis=['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Pregnancies']
for column in data1.columns:
    data1[column] = (data1[column] - data1[column].min()) / (data1[column].max() - data1[column].min()) 
data1.head()
```

I thought of working with the mean data alone, since i wanted to get data from the user. The data he enters would more or less be same to that of mean compared to se and worst case.

```python
x=data1.drop(['fractal_dimension_worst','symmetry_worst','concave points_worst','concavity_worst','compactness_worst','smoothness_worst','area_worst','perimeter_worst','texture_worst','radius_worst','fractal_dimension_se','symmetry_se','concave points_se','concavity_se','compactness_se','smoothness_se','area_se','perimeter_se','texture_se','radius_se'],axis=1)
x.head()
```

```python
y=data['diagnosis']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.1,stratify=y)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.svm import SVC
svm_model= SVC(kernel='rbf',gamma=8)
svm_model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
predictions= svm_model.predict(xtrain)
percentage=svm_model.score(xtrain,ytrain)
res=confusion_matrix(ytrain,predictions)
print("Training confusion matrix")
print(res)
predictions= svm_model.predict(xtest)
percentage=svm_model.score(xtest,ytest)
res=confusion_matrix(ytest,predictions)
print("validation confusion matrix")
print(res)
print(classification_report(ytest, predictions))
# check the accuracy on the training set
print('training accuracy = '+str(svm_model.score(xtrain, ytrain)*100))
print('testing accuracy = '+str(svm_model.score(xtest, ytest)*100))
```

The SVM performed well and produced 98% test accuracy, it produced 1 false negative and 0 false positive in test set which is a tremendous thing. the next is training of ann for the same data.

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
dl_model = Sequential() 

dl_model.add(Dense(256,  activation = 'relu' ,input_shape=([10]))) #input layer
dl_model.add(Dense(256,  activation = 'relu'))
dl_model.add(Dense(1,activation = 'sigmoid'))
dl_model.summary()
dl_model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' ,metrics = ['accuracy','Precision','Recall','AUC'])
```

```python
num_epochs = 70
history = dl_model.fit(xtrain ,
                    ytrain ,
                    epochs= num_epochs ,
                    steps_per_epoch=70,
                    validation_data=(xtest ,ytest))
```

```python
dl_model.save('breast_cancer_radiological_data-ANN.h5')
```

```python
dl_model.evaluate(xtest,ytest)
```
